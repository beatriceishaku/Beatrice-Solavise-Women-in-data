```python
import math

# Function to calculate area of a circle
def area_of_circle(radius):
    return math.pi * radius ** 2

# Function to calculate area of a rectangle
def area_of_rectangle(length, width):
    return length * width

radius = float(input("Enter the radius of the circle: "))
circle_area = area_of_circle(radius)
print(f"\nThe area of the circle with radius {radius} is: {circle_area:.2f}")

length = float(input("\nEnter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
rectangle_area = area_of_rectangle(length, width)
print(f"The area of the rectangle with length {length} and width {width} is: {rectangle_area:.2f}")

```
